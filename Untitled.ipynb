```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.lower().str.replace(' ', '_', regex=False)
```

```python
df=df.rename(columns={'purchase_amount_{usd}':'purhase_amount'})
```

```python
# إزالة كل الرموز غير الضرورية واستبدال الشرطات بمسافات
df.columns = (
    df.columns
    .str.replace("_", "")        # إزالة كل الـ Underscores
    .str.replace("\s+", " ", regex=True)  # تصحيح الفراغات
    .str.strip()                 # إزالة الفراغات في البداية والنهاية
)
```

```python
print(df.columns)
```

```python
labels ={'Young Adult','Adult','Middle_aged','Senior'}
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping ={
    'fortnightly':14,
     'Weekly':7,
    'Monthly':30,
    'Quarteryly':90,
    'Bi_weekly':14,
    'Annually':365,
    'Every 3 Monthly':90
}
df['purchase_frequency_days']=df['frequencyofpurchases'].map(frequency_mapping)
```

```python
print(df.columns.tolist())
```

```python
df=df.drop('promocodeused',axis=1)
```
